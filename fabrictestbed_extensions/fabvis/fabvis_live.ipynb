# FABRIC Slice Visualizer — Live

Visualize your real FABRIC slices interactively. Requires a configured FABRIC environment with valid credentials.

Two views are available:
1. **SliceVisualizer** — Cytoscape graph topology (nodes, components, networks)
2. **GeoVisualizer** — Geographic map with sites, nodes, and animated network paths

## Setup

Initialize FABlib and verify credentials.

```python
from fabrictestbed_extensions.fablib.fablib import FablibManager

fablib = FablibManager()
fablib.show_config()
```

---
## 1. Cytoscape Topology Viewer

Select slices from the dropdown and click **Load** to visualize.

**Interactions:**
- Click any **node** (box) → details panel shows cores, RAM, site, components
- Click a **component** (circle) → model, PCI address, interfaces
- Click a **network** (diamond) → type, subnet, connected interfaces
- Click an **edge** (line) → IP, VLAN, MAC, bandwidth
- Change **Layout** dropdown to rearrange the graph
- Click **Refresh** to reload data from FABRIC

```python
from fabrictestbed_extensions.fabvis import SliceVisualizer

vis = SliceVisualizer(fablib)
vis.show()
```

## 2. Geographic Map Viewer

Shows all FABRIC sites on a world map. Load slices to see nodes at their sites with animated network paths.

**Interactions:**
- Click a **blue site marker** → see site info and nodes deployed there
- Click a **colored node marker** → see node details (state, cores, RAM, etc.)
- Click an **animated ant path** → see the network connection details
- Toggle **Show all sites** checkbox to hide/show background sites
- Use scroll/pinch to zoom, drag to pan

```python
from fabrictestbed_extensions.fabvis import GeoVisualizer

geo = GeoVisualizer(fablib)
geo.show()
```
