# FABRIC Artifact Browser — Demo

This notebook demonstrates the **ArtifactBrowser**, an interactive tool for searching and downloading artifacts from the FABRIC Artifact Manager.

**What it does:**
- Fetches the full artifact catalog from the FABRIC Artifact Manager
- Displays artifacts as styled cards with title, description, tags, project, and authors
- **Scrollable list** — all cards live in a fixed-height scrollable area so 100+ artifacts don't crowd the page
- **Live filtering** by title and tag keywords
- **Version selector** dropdown per artifact
- **One-click download** with automatic tar extraction
- **Refresh** button to reload the catalog

**Sections:**
1. **Quick Start** — Launch the browser with a live FablibManager
2. **Custom Download Directory** — Change where artifacts are saved
3. **Mock Demo** — Try the UI without a live FABRIC connection

---
## 1. Quick Start

Launch the artifact browser using a live FablibManager connection.

> **Prerequisites:** You must have a valid FABRIC configuration (run the **ConfigureGUI** or `configure_and_validate` notebook first).

````python
from fabrictestbed_extensions.fablib.fablib import FablibManager
from fabrictestbed_extensions.fabvis import ArtifactBrowser

fablib = FablibManager()

browser = ArtifactBrowser(fablib)
browser.show()
````

## Quick Reference

```python
from fabrictestbed_extensions.fablib.fablib import FablibManager
from fabrictestbed_extensions.fabvis import ArtifactBrowser

fablib = FablibManager()

# Default (downloads to ~/work)
browser = ArtifactBrowser(fablib)
browser.show()

# Custom download directory
browser = ArtifactBrowser(fablib, download_dir="/home/fabric/artifacts")
browser.show()
```

**UI Controls:**
| Control | Action |
|---------|--------|
| **Title filter** | Live-filter artifacts by title keyword |
| **Tag filter** | Live-filter artifacts by tag keyword |
| **Refresh** | Reload the artifact catalog from the server |
| **Version dropdown** | Select which version to download (per card) |
| **Download** | Download and auto-extract the selected version |
