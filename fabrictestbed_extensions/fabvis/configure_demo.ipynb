# FABRIC Environment Configuration — Demo

This notebook demonstrates the **ConfigureGUI**, an interactive tool for setting up your FABRIC environment.

**Features:**
- **Auto-loads from existing config** — if `fabric_rc` already exists, all fields are pre-populated
- **Auto-loads projects** — projects are fetched automatically on startup (if the token file is valid)
- **Token file at top** — highlighted with a red indicator if missing or invalid
- **Project dropdown** — select from your projects, or click **Manual** to paste a UUID directly
- **Auto-filled bastion username** — populated from your FABRIC user info
- **File browser** — click *Browse* next to any file path to navigate and select files
- **Log level dropdown**, **log file**, and **SSH command line** — all configurable
- **Sites to Avoid** — clickable toggle buttons for every FABRIC site
- **Preview RC File** — see exactly what will be saved before committing
- **Validate & Configure** / **Save Config** / **Reset Defaults**

**Sections:**
1. **Quick Start** — Launch the GUI
2. **Custom Config Path** — Use a non-standard `fabric_rc` location
3. **Programmatic Access** — Read/set field values from code

---
## 1. Quick Start

Launch the configuration GUI. If you already have a `fabric_rc` file, all fields are pre-populated automatically. Projects are loaded from FABRIC on startup.

**If this is your first time:**
1. Set the **Token File** path — the token indicator turns green when valid, and projects load automatically
2. Select your **project** from the dropdown (or click **Manual** to paste a UUID)
3. Adjust file paths if needed — use the **Browse** buttons
4. Toggle any sites to **avoid**
5. Click **Preview RC File** to see what will be saved
6. Click **Validate & Configure** to test everything
7. Click **Save Config** to persist your settings

> **Tip:** If the token file is wrong, you'll see a red indicator and the project dropdown will say "failed to load". Fix the token path and projects will reload automatically.

````python
from fabrictestbed_extensions.fabvis import ConfigureGUI

gui = ConfigureGUI()
gui.show()
````

## 3. Programmatic Access

After the GUI is constructed, you can read or set field values programmatically via the `_fields` dictionary.

````python
# Read current field values
print("Available fields:", list(gui._fields.keys()))
print()
for key, widget in gui._fields.items():
    print(f"  {key}: {widget.value!r}")
````

## Quick Reference

```python
from fabrictestbed_extensions.fabvis import ConfigureGUI

# Default — loads existing fabric_rc if present
gui = ConfigureGUI()
gui.show()

# Custom fabric_rc path
gui = ConfigureGUI(fabric_rc="/home/user/my_fabric_rc")
gui.show()
```

**Identity Section:**

| Control | Behavior |
|---------|----------|
| **Token File** | Path to your FABRIC token. Green check = valid, red X = missing. Changing this auto-reloads projects. |
| **Project** | Dropdown populated from FABRIC. Click **Manual** to toggle to a text field for pasting a UUID directly. Click **Reload** to re-fetch. |
| **Bastion Username** | Auto-filled from your FABRIC user info when projects load. |

**Action Buttons:**

| Button | Action |
|--------|--------|
| **Validate & Configure** | Creates FablibManager, shows config, runs `verify_and_configure()` |
| **Save Config** | Writes settings to `fabric_rc` |
| **Preview RC File** | Shows the exact `export FABRIC_XXX=...` lines that will be saved |
| **Reset Defaults** | Restores all fields to default values |
